```python
def move_complete_cluster(clusters, dex='univ2'):
    cluster_path = eval(f'path.{dex}_cluster_path')
    complete_cluster_path = os.path.join(eval(f'path.{dex}_processed_path'), "complete_cluster")
    for cluster_id in clusters:
        file_name = f'cluster_{cluster_id}.csv'
        from_path = os.path.join(cluster_path, file_name)
        to_path = os.path.join(complete_cluster_path, file_name)
        shutil.copyfile(from_path, to_path)
```
